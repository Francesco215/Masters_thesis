```python
import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np
import texpy as mz
from src import *
```

```python
def E(n,phi):
    return (n+phi)**2
```

```python
cmap=plt.get_cmap('viridis')
colors=cmap(np.linspace(0,1,10))

"""plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
    "font.size":12,
})"""

for i,n in enumerate(np.arange(-4,5)):
    phi=np.linspace(-40,40,100)
    plt.plot(phi/10,E(n*10,phi),color=colors[i],label=r'$\mathrm{}$'.format(n),linewidth=2.5)

plt.ylim(0,400)
plt.xlim(-4,4)
plt.ylabel(r'$E_n$',fontsize=17)
plt.xlabel(r'$\Phi/\Phi_0$',fontsize=15)
plt.yticks([])
plt.legend(loc='right', bbox_to_anchor=(1.23,0.5), shadow=True,title='n')
plt.savefig('spectral_flow.svg',format='svg',bbox_inches='tight')
plt.show()
```
