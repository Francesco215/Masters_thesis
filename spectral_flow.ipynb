```python
import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np
import texpy as mz
from src import *
```

```python
def E(n,phi):
    return (n+phi)**2
```

```python
cmap=plt.get_cmap('viridis')
colors=cmap(np.linspace(0,1,10))

"""plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
    "font.size":12,
})"""

phi=np.linspace(-40,40,100)
for i,n in enumerate(np.arange(-4,5)):
    plt.plot(phi/10,E(n*10,phi),color=colors[i],label=str(n),linewidth=2.5)

plt.ylim(0,400)
plt.xlim(-4,4)
plt.ylabel('$E_n$',fontsize=17)
plt.xlabel('$\Phi/\Phi_0$',fontsize=15)
plt.yticks([])
plt.legend(loc='right', bbox_to_anchor=(1.23,0.5), shadow=True,title='n')
plt.savefig('spectral_flow.svg',format='svg',bbox_inches='tight')
plt.show()
```

```python
def band(k,n=0,t0=.1):
    E_n=n**2
    return E_n-(-1)**(n)*t0*np.exp(-n/(2*np.pi))*(np.cos(k*(np.pi))-2)
```

```python
k=np.linspace(-1,1,100)

for n in range(0,3):
    plt.plot(k,band(k,n,t0=1))
```
